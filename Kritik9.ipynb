```python
import sympy as sp
x, y = sp.symbols('x y')
```

```python
#Part a
f = sp.exp(x)*sp.sin(y) + y**3

fpartialx=sp.diff(f, x)
fpartialy=sp.diff(f, y)

print(fpartialx)
print(fpartialy)
```

```python
#Part b
g = (x**2)*y + x*(y**2)

gpartialx=sp.diff(g, x)
gpartialy=sp.diff(g, y)

def f(x, y):
    return (2*x*y + y**2)

def g(x, y):
    return (x**2 + 2*x*y)

f(1, -1)
g(1, -1)

gradient= sp.sqrt(f(1, -1)**2 + g(1, -1)**2)
print(f"The gradient vector has a magnitude of {gradient}")

```

```python
#Part c
h = sp.log((x**2) + (y**2))

hpartial1x=sp.diff(h, x)
hpartial1y=sp.diff(h, y)

h=hpartial1x
hpartial2x=sp.diff(h, x)
hpartialxy=sp.diff(h, y)

h=hpartial1y
hpartial2y=sp.diff(h, y)
hpartialyx=sp.diff(h, x)

print(hpartial2x)
print(hpartialxy)
print(hpartial2y)
print(hpartialyx)

#Notice that the second and fourth outputs are identical! This is because of clairaut's theorem where if the second
#partial derivatives are continues, then the mixed second partial derivatives are symmetrical. 
```
