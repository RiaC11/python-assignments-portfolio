```python
import numpy as np
import matplotlib.pyplot as plt 

def normal_density(mean, variance, x):
    partone=1/np.sqrt(2*np.pi*variance)
    parttwo=np.exp((-(x-mean)**2)/(2*variance))
    return partone*parttwo

print("Done!")
```

```python
#Question 3
mean=2
variance=2

x=np.linspace(-10, 10, 100)
y=normal_density(mean, variance, x)

plt.plot(x, y)
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
#Question 3 continued
def integration(mean, variance, a, b): #Use the numerical integration technique of Riemann sums
    
    #define the parameters
    n=10000 # number of rectangles

    #define the x and y variables  
    x = np.linspace(a,b,n+1) #creates an array of n+1 equally spaced x-values between a and b
    y = normal_density(mean, variance, x) #define y=f(x) or the height of each rectangle - so this depends on what the function is

    #calculate the width of each rectangle
    w = (b-a)/n

    #compute the areas of all possible rectangles
    area = y*w 

    #compute the left-sum (a[0]+a[1]+...+a[n-1])
    ls = sum(area[0:n])
    print("The left sum is:",ls)

    return ls

result = integration(171, 7.1, 162, 190)

#Note that this result makes sense because 99% of the values will occur between 3 standard deviations above the mean (resulting in 192) and 3 standard deviations below (resulting 171).
#Note that the standard deviation is the square root of the mean so 7.1. 
```

```python
#Question 4c. Note that 4a and 4b were done by hand.

#When the question asks to find the average dosage, this means we need to find the expected value
#So we have to do the integral of the dosage times the normal density function

n=10000 #number of rectangles

def average_dosage(mean, variance, a, b):
    x=np.linspace(a, b, n+1) #creates the array of n+1 equally spaced x-values between a and b
    y=(2.38*x**2)*normal_density(mean, variance, x) #define y=f(x) or the height of each rectangle
  
    w=(b-a)/n #this is the width of the rectangle
    area=y*w #this is the area of each rectangle 
    answer=sum(area[0:n]) #this sums all the areas
    return answer

average_dosage(171, 7.1, 162, 190)
```
